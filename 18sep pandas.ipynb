```
#method 5
#Creating a Dataframe using the zip() function

#list1
name=['Alice','Bob','Charlie','David','Karl','Amile','Daminan']

#list2
age=[25,30,22,28,19,22,27]

#list3
score=[85,70,90,80,75,77,82]

#list4
city=['New York','Los Angeles','Frankfurt','London','Oslo','Barcelona','Frankfurt']

#using zip to merge the four lists
list_of_tuples=list(zip(name,age,score,city))

df=pd.DataFrame(list_of_tuples, columns=['Name','Age','Score','City'])

df
```

```
#1. Selection by column name (single or list)
selected_df_1=df['Name'] #selecting a single column by name
selected_df_2=df[['Name','Score']] #selecting a list of column by name
print(selected_df_1)
print()
print(selected_df_2)

print()
print('*************************************************')
print()

#2.Selecting by positional indexing
selected_df_3=df[0:2] #selecting first two columns(zero-based indexing)
selected_df_4=df.iloc[:,1:3] #selecting column from index 1 inclusive to column 3
print(selected_df_3)
print()
print(selected_df_4)

print()
print('*************************************************')
print()

#3.Selecting with. Loc(Label-based)
selected_df_6= df.loc[:,['Name','City']] #selecting column from name and city
selected_df_6
```

```
##sorthing row within pandas  dataframe
import pandas as pd

raw_data={"First_name":["Sheldon","Raj","Leonard","Howard","Amy"],
         "Last_name":["Copper","Koothrappali","Hofstadter","Wolwitz","Flower"],
         "age":[42,38,36,41,35],
         "Comedy_Score":[9,7,8,8,5],
         "Rating_Score":[25,25,49,62,70]}

df=pd.DataFrame(raw_data, columns=["First_name","Last_name","age","Comedy_Score","Rating_Score"])
print(df)
```

```
print(df.sort_values(by="age", ascending=0)) #decending

print()
print('*************************************************')
print()

print(df.sort_values(by="age", ascending=1)) #ascending

print()
print('*************************************************')
print()

print(df.sort_values(by=["age","Comedy_Score"]))
```

```
#group on a single column(age)
group_g=df.groupby(["age"])
print(group_g)
print(group_g.size())

print()
print('*************************************************')
print()

group_g=df.groupby(["age","Comedy_Score"])
print(group_g)
print(group_g.size())
```

```
#How to find the largest and Lowest values in padans DataFrame
print("Index of largest values: ");print(df['Rating_Score'].idxmax())
print("Index of lowest values: ");print(df['Rating_Score'].idxmin())

print("Mean of values: ");print(df['Rating_Score'].mean())
print("Median of values: ");print(df['Rating_Score'].median())

```
